```python
df.Handcap.replace({2:1,3:1,4:1},inplace=True)
df.Handcap.unique()
```

```python
df = df.rename(columns={'Hipertension': 'Hypertension', 'Handcap': 'Handicap', 'No-show': 'Show'})
df.columns
```

```python
df.head()
```

```python
df.Show.replace({'No': 'Yes', 'Yes' : 'No'}, inplace= True)
df.head()
```

```python
df['DaysLate'] = (df.AppointmentDay - df.ScheduledDay).dt.days 
df.head(1)
```

```python
df.DaysLate.min()
```

```python
df.query('DaysLate <= 0 & Show == "No"')
```

```python
df['Show'] = df.query('DaysLate <= 0 and Show=="No"').replace({'No':'Yes'},inplace=True)
df.query('DaysLate <= 0 and Show=="No"')
```
